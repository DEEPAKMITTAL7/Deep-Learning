```python
# Importing libararies for neural network

import tensorflow as tf
from tensorflow import keras

# Importing libararies for data manipulation

import pandas as pd
import numpy as np

# Importing libraries for data visualization

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

```python
df['age'] = df['age']/100
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[['age', 'affordibility']], df['bought_insurance'], test_size=0.2, random_state=0)
```

```python
model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(2,), activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(x_train, y_train, epochs=1500)
```

```python
model.evaluate(x_test, y_test)
```

```python
def prediction_function(age, affordability):
    weighted_sum = coef[0]*age + coef[1]*affordability  + intercept
    return sigmoid(weighted_sum)

prediction_function(0.28, 1)
```

```python
def sigmoid_numpy(z):
    return 1 / (1 + np.exp(-z))

sigmoid_numpy(np.array([100, 0, 1]))
```

```python
def log_loss(y_true, y_predicted):
    y_predicted = [ np.abs((1e-15)-i) if i in [0, 1] else i for i in y_predicted ]
    y_predicted = np.array(y_predicted)
    return -np.mean(  (y_true*np.log(y_predicted))  + (1-y_true) * (np.log(1-y_predicted))  )

log_loss(np.array([1,1,1,1]), np.array([0,1,0,1]))
```

```python
def gradient_descent(x, y, epochs, log_los):
    w1 = w2 = 1
    bias = 0

    learning_rate = 0.5
    n = len(x['age'])

    for i in range(epochs):

        weighted_sum = w1*x['age'] + w2*x['affordibility'] + bias
        y_predicted = sigmoid_numpy(weighted_sum)
        loss = log_loss((y), (y_predicted) )

        w1_ = (1/n)*np.dot(np.transpose(x['age']), (y_predicted-y))
        w2_ = (1/n)*np.dot(np.transpose(x['affordibility']), (y_predicted-y))

        bias_ = np.mean(y_predicted-y)

        w1 = w1 - learning_rate * w1_
        w2 = w2 - learning_rate * w2_

        bias = bias - learning_rate * bias_

        print(f"iteration:{i}, cost: {loss}, w1:{w1}, w2:{w2}, bias:{bias}")

        if loss <= log_los:
            break

gradient_descent(x=x_train, y=y_train, epochs=1500, log_los=0.5923)
```
