```python
# Importing libararies for neural network

import tensorflow as tf
from tensorflow import keras

# Importing libararies for data manipulation

import pandas as pd
import numpy as np

# Importing libraries for data visualization

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('sonar_dataset.csv', header=None)
df.sample(5)
```

```python
df[60].value_counts()
```

```python
x = df.drop(60, axis= 1)
y = df[60]
```

```python
x
```

```python
y = pd.get_dummies(y, drop_first=True)
y
```

```python
y.replace({'True':1, 'False':0}, inplace=True)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_shape=(60,), activation= 'relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss= 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(x_train, y_train, epochs=100, batch_size= 8)
```

```python
model.evaluate(x_test, y_test)
```

```python
y_predicted = model.predict(x_test).reshape(-1)
y_predicted = np.round(y_predicted)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predicted))
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_shape=(60,), activation= 'relu'),
    keras.layers.Dropout(0.5),               # drops 50% neurons
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss= 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(x_train, y_train, epochs=100, batch_size= 8)
```

```python
model.evaluate(x_test, y_test)
```
